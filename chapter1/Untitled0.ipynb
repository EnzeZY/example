```python
library(shiny)
ui = pageWithSidebar(
  headerPanel(""),
  sidebarPanel(
    selectInput("x", "", names(mtcars), "wt")  
  ),
  mainPanel(
    chartOutput("r1", "polycharts")
  )
)
```

```python
server = function(input, output){
  output$r1 <- renderChart2({
    rPlot(x = input$x, y = "mpg", data = mtcars, type = 'point')
  })
}
```

```python
runApp(list(ui = ui, server = server))
```
